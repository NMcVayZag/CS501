### Classes and Objects (Ch 4.1)
* The essesense of OOP development is the process of designing and implementing our own classes to suit our specific needs.
* A class is the blueprint of an object.
* An object has a "state" that is defined by the values of the attributes associated with the object
* An object also has behaviors which are defined by the operations(methods) associated with that object
* an object should be self governing
- Scope of a variable
	- the area within a program that the variable can be referenced
	- A scope of a variable or attribute is specified when initialized- an attribute can be public and can be access by areas outside the class declaration and it can be private which can only be referenced from within the class
### Anatomy of a Class (Ch 4.2)
- Instance Data
	- refers to data or variables that are created everytime a new instance of a class is created. Each new instance of a class creates new memory slots for attributes related to that class. So that each object of a class can have its own "State"
- UML class diagrams
	- UML or Unified Modeling Language is the most popular notation for representing the design of an object oriented program.
	- In a UML diagram each class is represented as a rectangle possily containing three sections to show the class name, its attributes, and its methods 
### Encapsulation (Ch 4.3)
- ENCAPSULATION: is the characteristic in objects we strive for when designing OOPs - basically we want to make it difficult for code outside a class to "reach in" and change the value of a variable that is declared inside of that class(self governing principle). An object should only interact with the rest of the program through its methods()
	- in java, object Encapsulation is acheived through **modifiers**
		- Modifiers are java reserved words that are used to specify particular characteristics of a programming language construct. ex. Final int someInt;
- Visibility modifiers
	- visibility modifiers control access to members of a class. "public" and "private" are visibility modifiers that can be applied to variables and methods of a class.
		- **if a member of a class has public visibility, it can be directly referenced from outside the object. This violates the practice of abstraction allowing code outside the class definition to access and modify its variables. Therefore anything with instance data should be defined with private visibility. However, most class methods should be kept public so they can be called from outside the class. The only methods that should be kept private are the ones that assist other methods within the class - often referred to as "support methods". denoted in UML with "+"**
		- if a member of a class has private visibility, it can be used anywhere inside the class definition but cannot be referenced externally. denoted in UML with "-"
- Accessors and Mutators (Getters and Setters)
	- Acessors(getter): are methods that provide read-only access to a partucular value. Often used by public class methods that allow for the reading of private instance data variables. 
	- Mutators(setter): are methods that provide read/write access to a particular value/variable. Often used by public class methods allowing for the reading/writing of private instance data variables.

### Anatomy of a Method (Ch 4.4)
* all methods of a java program must be part of a class
* the header of a method includes the type of the return value, method name, and a list of parameters the method accepts
- return statement
	- the statement at the end of a method that is an expression which dictates the value to be returned by the method. it is best practice to only have one per method
- Parameters and Arguments
	- a parameter is a value that is passed into a method when its invoked. the parameter list in the header of a method specifies the type of the values that are passed and the names by which the method will refer to those values.
		- the parameters passed into a method when its being called are known as "actual parameters" whereas those named in the method header are referred to as "formal parameters"- **the values/types are copied over**.
- Local Data
	- refers to data that is declared inside a method making it "local data" not instance data. Its called local as its scope is bound to the method itself.
### Constructors Revisited (Ch 4.5)
- A constructer differs from a regular method in two ways: 
	- first the name of a constructer is the same as its class
	- second a constructor cannot return a value and doesn't have a specified return type in the method header. Putting a type on a constructer turns it into a regular method - leading to an error
	
### Boolean Expressions (Ch 5.1)
- Equality and Relational Operators 
	- Equality operators
		- tell whether two values are equal or not equal
	- relational operators
		- tell us whether an variable is smaller or larger than another
- Logical Operators
	- There are three logical operators that help us find out if something is true or not.
	- ! is the NOT operator and is the logical compliment(opposite) of a boolean value. !ON would mean its not on where ON is true.
	- AND(&&) lets us create a boolean value if two conditions are met
	- OR(||) lets us create a boolean value if one of the conditions are met.
### The if Statement (Ch 5.2)
- if
	- an if statement consists of the reserved word if followed by a boolean expression, followed by a statement that executes some code if the boolean is found to be true.
- if-else
	- is a appended if statement that is only executed if the previous if statement condition was not met- then this if statement will be run.
- Using block statements
	- a block statement is a collection of statemnts enclosed in braces 
- Nested if statements
	- occurs when an if statement is placed in the statement of another if statement.
### Comparing Data (Ch 5.3)
- Comparing floats (this includes doubles)
	- two floating point values are equal according to the "==" operator only if all the binary digits of there underlying representations match. This is rarely the case so it is more ideal to evaluate the absolute value of the difference of those float values
- Comparing characters
	- characters are evaluated in terms of their unicode values so that a is less than b
- Comparing objects
	- String object equality can be easily be evaluated using "==" to check if they reference the same object("alias"), or str.equals() method to check if the characters are the same.
### Recursive Thinking (Ch 12.1)
 - Infinite Recursion
	- can occur if a method endlessly calls itself as the only option which creates an infinite loop.
- Recursion in Math
	- recursion is used in math for factorials where N! is defined for any positive integer N as the product of all integers between 1 and N. where the factorial function is defined in terms of the factorial function
### Recursive Programming (Ch 12.2)
- Tracing recursion
	- is process of following a recursive method though its steps at every call keeping track of variable changes in order to understand them and be capable of debugging them.
- Recursion vs Iteration
	- A programmer must know that all recursion problems can be solved through iteration however some problems can use recursion to create a cleaner more efficient solution

- Direct vs Indirect
	- direct recursion occurs when a method invokes itself where as Indirect recursion occurs when a method inovkes another method which eventually results in the original method being called again.
