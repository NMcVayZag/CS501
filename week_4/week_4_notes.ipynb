### The while Statement (Ch 5.4)
- How it works
	- A wile statement is a loop that evaluates a boolean condition just as an if statement does and executes a statement(called the body of the loop) if that condition is true. It only breaks when that condition becomes false.
- Sentinal value
	- A sentinal value is one that is used to signal a wish to exit a program or loop. A sentinal value must always be outside the normal ramge of values entered
- Infinite loops
	- can occur when the program is designed incorrectly to where a loop's body will execute forever. A well designed while loop will have a intentional catch mechanism included such as an iteration limit or a user input stoppage
- Nested loops
	- When a loop is declared in the body of another loop.
- The break and continue Statements
	- break: 
		- a break statement will immedietly cause the program to exit the current loop(bad practice). It is also commonly used when writing switch statments.
		- Using a break in a loop body is actually bad practice as it causes the program flow to jump from one place to another. An equivalent loop can always be written without a break statement.
	- continue: 
		- a continue statement is similar to a break but the loop is evaluated again and if true the loop body will execute once again(bad practice).
		- once again using a continue statement in a loop is fully avoidable and for the same reasons should be avoided
### Iterators (Ch 5.5)
- Definition:
	 - an iterator is an object that has methods that allow you to process a collection of items one at a time. An iterator lets you step through each item and interact with it as needed.
	 - every iterator object has a method called "hasNext" that returns a boolean value indicating if there is at least one more item to process. You can use hasNext() as a condition of a loop to control the processing of each item. It also has a method called "next()" to retrieve the next item in the collection to process
	 CODE EXAMPLE: 
								import java.util.*;
								public class IteratorExample {
								public static void main(String[] args) {
									List<String> list = new ArrayList<>();
									list.add("Apple");
									list.add("Banana");
									list.add("Cherry");

									Iterator<String> iterator = list.iterator(); // Get an iterator for the list
									
									while (iterator.hasNext()) {
										String fruit = iterator.next();  // Retrieve the next element
										System.out.println(fruit);       // Print the current element
									}
								}
							}

### The ArrayList class (Ch 5.6)(singular data type storage)
 - Defining
	- An ArrayList object stores a list of objects and allows you to refer to each one by an integer index value.
	- An ArrayList an a array list has a generic type(E) which is determined when an ArrayList object is created -meaning that when you create an ArrayList you must specify the type of data or varibles that will be stored within it. So an array of strings is initialized with ArrayList<String> or an array of books is ArrayList<Book>

- Important ArrayList methods
	- ArrayList<TYPE>() - creates an initially empty list
	- boolean add(TYPE obj) - inserts the specified object to the end of this list
	- void add(int index, TYPE obj) - inserts the specified object into this specified index
	- void clear() - removes all elements from the list
	- TYPE remove(int index) - removes the element at the specified index in this list and returns it
	- TYPE get(int index) - reutrns the object at the specified index from the list without removing it
	- int indexOf(Object obj) - returns the index of the first occurance of the specified object 
	- boolean contains(Object obj) - returns true if this list contains the specified object
	- boolean isEmpty() - returns true if list contains no elements
	- int size() - returns the number of elements in the list
### The switch Statement (Ch 6.1)
- Defining
	- the switch statement causes the executing program to follow one of the several paths based on a single value. So you could have multiple statements labeled "case 'A':" or "case 'B':" and whatever the switch is defined as the program will go and execute the code under that case. it is good practice to define a default case in the situation where no specified case is reached by the expression

### The for Statement (Ch 6.4)
- Defining 
	- the for statement is another loop designed to executing a body of a loop a specific numner of times that can be determined before the loop is executed.
	- ex:
		- for (int count=1; count<=5; count++){
			System.out.println(count);
		}
- Structure
	- the header of a for loop contains three parts separeated by semicolons. 
		- Before the loop beigns, the first section of the header known as "initialization" is executed.
		- the second part of the header is the boolean condition which is evaluated before the loop body is executed. 
		- if the condition in second section is true then after the body of the loop is executed the third statment in the header known as the increment is executed.
- The for-each loop
	- the for each loop is a variation of the for statement is particularly ideal in situations that involve iterators(objects that have hasNext() or next() methods) since for this case we can use a simplified syntax an example of this would be iterating through an ArrayList<Book> named library
		- ex:
			- for (Book myBook : library)
				System.out.println(myBook)
- Comparing loops
	- we use a for loop when the number of iterations neccessary is fixed or can be easily calculated. 
	- we use while loops when we are not sure how many iterations a process may take to reach the desired outcome.
